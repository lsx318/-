```python
# 加载sklearn自带乳腺癌数据集
X, y = load_breast_cancer(return_X_y=True)

# 将特征数据转换为pandas DataFrame
import pandas as pd
X = pd.DataFrame(X)

# 查看数据集的形状
print(X.shape)

# 查看前几行数据
print(X.head(5))

# 查看数据类型和缺失值
print(X.info())

# 描述性统计分析
print(X.describe())

# 特征分布
import matplotlib.pyplot as plt
import seaborn as sns

# 为了简洁，这里只展示前两个特征的直方图
sns.histplot(X.iloc[:, 0], kde=True)
plt.show()
sns.histplot(X.iloc[:, 1], kde=True)
plt.show()

# 相关性分析
corr = X.corr()
print(corr)

# 可视化相关性矩阵
sns.heatmap(corr, annot=True)
plt.show()
```

划分数据集和训练集

```python
X_train,X_test,y_train,y_test=train_test_split(
    X,y,
    train_size=0.875,test_size=0.125,random_state=188
)
```

创建逻辑斯蒂回归模型并训练

```python
from sklearn.linear_model import LogisticRegression

# 使用lr类，初始化模型
clf = LogisticRegression(
    penalty="l2",  # 设置正确的penalty参数值
    C=1.0,
    random_state=None,
    solver="lbfgs",
    max_iter=3000,
    multi_class="ovr",
    verbose=0,
)

# 进行拟合
clf.fit(X_train, y_train)
```

$$
\text{LogisticRegression}(max\_iter=3000, multi\_class='ovr')
$$

```python
# 预测测试集
y_pred = clf.predict(X_test)

# 评估模型性能

print("Accuracy:\n", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
print("Paramas:\n",clf.coef_,clf.intercept_)
```
